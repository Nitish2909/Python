DISTRIBUTION PLOT AND ITS TYPES

```python
# 1. Distplot/histplot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
df

# in this dataset total_bill and tips are quantitative data, others are categorical or qualitative data
```

```python
# 1. sns.histplot(data,bins= value,kde=Thrue)
# default size of bins is 20

sns.histplot(df["total_bill"],bins=20,kde=True)
# kde = kernel density estimater.this shows skewness in your data
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df["tip"],bins=20,kde=True)
```

2. jointplot

```python
#jointplot => A jointplot is used to visualize the relationship between two numerical variables along with their individual distributions.
# syntax:
# sns.jointplot(x= x_axis,y=yaxis,data,kind="scatter")
# kind = ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid'] choose one from this for kind

sns.jointplot(x= "total_bill",y="tip",data=df,kind="scatter")
```

3. pairplot

```python
# pairplot =>A pairplot is used to visualize relationships between multiple numerical variables at once and also shows individual distributions on the diagonal.
#In short: Pairplot = many jointplots combined into one grid.

# Syntax:
# sns.pairplot(data,hue="category",palette="rainbow")

sns.pairplot(df,hue="sex",palette="rainbow")
```

4. rugplot

```python
# rugplot
# rugplot(data)

sns.rugplot(df)
```

### boxplot

A boxplot (box-and-whisker plot) is a statistical plot used to show the distribution of numerical data(Quantitative data) and detect outliers, often across categories.
on x-axis= quantitative data
on y-axis = qualitative data

```python
sns.boxplot(x='tip',y='day',data=df)
```

violintplot:
A violinplot is a statistical plot that shows the distribution of numerical data across categories, combining a boxplot and a KDE (density plot).


```python
sns.violinplot(x='day',y='total_bill',data=df)
```

swarmplot:
A swarm plot is a categorical plot that shows all individual data points for each category, arranged so they don’t overlap.

```python
sns.swarmplot(x='day',y='total_bill',data=df)
```

stripplot:
A stripplot is a categorical plot that shows individual data points for each category, with points placed along an axis.
Unlike swarmplot, points can overlap.

```python
sns.stripplot(x='day',y="tip",data=df)
```

**MATRIX PLOT**

A matrix plot (in Seaborn/EDA context) is used to visualize data arranged in a matrix form, most commonly to show correlations between variables using colors.
In simple words:
Matrix plot = relationship strength shown.

Types of Matrix Plot:
1. heatmap

2. cluster map

3. pivot table heat map

```python
import seaborn as sns;
import matplotlib.pyplot as plt;

tips= sns.load_dataset("tips")
tips
```

```python
tipsCorrelation = tips[['total_bill','tip','size']]
tipsCorrelation
```

```python
tipsCorrelation.corr()
```

```python
sns.heatmap(tipsCorrelation.corr(),annot=True)
```

cluster map

```python
sns.clustermap(tipsCorrelation.corr())
```

pivottable heat map

```python
flights= sns.load_dataset("flights")
flights
```

Regression Plots

A regression plot (reg plot) is used to visualize the relationship between two numerical variables along with a best-fit regression line.

In simple words:

Regression plot = scatter points + trend line

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips
```

```python
# what you want to predict you write on y-axix
# the data that is used in prediction is write on x-axis
sns.lmplot(x="total_bill",y="tip",data=tips)
```
