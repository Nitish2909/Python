```python
import pandas as pd
data = {
    "Name" :["Nitish","Rakesh","Rohan","Raju","Sham","Baburao"],
    "age" :[20,21,22,34,23,45],
    "Salary" : [20000,30000,80000,45000,55000,88000]
}
df = pd.DataFrame(data)
print(df)
```

```python
df.to_csv('simple_data.csv',index=False) # Export file
```
