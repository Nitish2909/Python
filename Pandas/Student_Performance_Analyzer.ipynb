```python
import pandas as pd
```

```python
df=pd.read_excel('StudentData.xlsx/StudentData.xlsx')
```

```python
df
```

```python
# It make a column that name is "Total"
# It adds total 5 subject Marks of each students and present in their row in "Total" Name column

df['Total'] = df['English'] + df['Maths'] + df['Hindi'] + df['Social-Science'] + df['Science']
```

```python
df
```

```python
# Average of Marks
```

```python
df['Average'] = df['Total']/5
df
```

```python
# Find Highest Marks from Each Subject
```

```python
print("Highest Marks in English :", df['English'].max())
print("Highest Marks in Maths :", df['Maths'].max())
print("Highest Marks in Hindi :", df['Hindi'].max())
print("Highest Marks in Social-Science :", df['Social-Science'].max())
print("Highest Marks in Science :", df['Science'].max())
```

```python
# Find Toppers from each subject with their Name
```

```python
print("English Topper :")
print(df.loc[df['English'].idxmax() , ['Name', 'English']])

print("\n Maths Topper :")
print(df.loc[df['Maths'].idxmax() , ['Name', 'Maths']])

print("\n Hindi Topper :")
print(df.loc[df['Hindi'].idxmax() , ['Name', 'Hindi']])

print("\n Social-Science Topper :")
print(df.loc[df['Social-Science'].idxmax() , ['Name', 'Social-Science']])

print("\n Science Topper :")
print(df.loc[df['Science'].idxmax() , ['Name', 'Science']])
```

```python
# find Grade from Average
```

```python
def grade(avg):
  if avg >= 80:
    return 'A'
  elif avg >= 70:
    return 'B'
  else:
    return 'C'


df['Grade'] = df['Average'].apply(grade)
```

```python
df
```

```python
# Find Topper
```

```python
topper = df[df['Total'] == df['Total'].max()]
print("topper : \n",topper )
```

```python
weak_student = df[df['Average']<65]
print("Needs Improvement : \n",weak_student)
```
