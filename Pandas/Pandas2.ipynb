```python
# Concatenate & Merge DataFrames (JOINs)
```

```python
import pandas as pd

data={
    "Id":[0,1,2,3],
    "Name":["Nitish","Rohan","Manish","Sohan"],
    "Marks":[80,87,45,78]

}
df1 = pd.DataFrame(data)
print(df1)
```

```python
import pandas as pd

data={
    "Id":[0,1,2,3],
    "Name":["Raju","Sham","Baburao","Soham"],
    "Marks":[83,85,60,70]

}
df2 = pd.DataFrame(data)
print(df2)
```

```python
# using concatenate
```

```python
pd.concat([df1,df2],axis=0)    # row level / vertical /top on top
```

```python
pd.concat([df1,df2],axis=1)  # Horizontal / column level / side on side
```

```python
# Merge (JOIN)
```

```python
pd.merge(df1,df2, how='inner', on='Id')
```
